```python
deliveries=deliveries.replace({
    'Delhi Daredevils': 'Delhi Capitals',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
})
```

```python
deliveries.to_csv("deliveries_cleaned.csv", index=False)
```

```python
##1. Top 10 Run Scorers
```

```python
top_batsman=(
    deliveries.groupby('batsman')['batsman_runs']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
    
```

```python
top_batsman
```

```python
##2. Top 10 Wicket Takers
```

```python
top_bowlers=(
    deliveries.groupby('bowler')['is_wicket']
    .sum()
    .sort_values(ascending=False)
    .head()
)
```

```python
top_bowlers
```

```python
## Most Sixes
```

```python
sixes = (
    deliveries[deliveries['batsman_runs'] == 6]
    .groupby('batsman')
    .size()
    .sort_values(ascending=False)
    .head(10)
)

sixes
```

```python
### Most Fours
```

```python
Fours=(
    deliveries[deliveries['batsman_runs']==4]
    .groupby('batsman')
    .size()
    .sort_values(ascending=False)
    .head(10)
)

Fours
```
